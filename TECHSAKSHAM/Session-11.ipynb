```python
import pandas as pd
import seaborn as sns
dataset=pd.read_csv('data.csv')
```

## STATISTICAL RELATIONSHIP - SCATTER PLOT

```python
sns.relplot(x='hours',y='marks',hue='age',data=dataset)
```

```python
sns.relplot(x='hours',y='marks',hue='internet',data=dataset)
```

## STATISTICAL RELATIONSHIP - LINE

```python
sns.relplot(x='hours',y='marks',hue='age',data=dataset,kind='line',style='internet')
```

## STATISTICAL RELATIONSHIP - MULTIPLE

```python
sns.relplot(x='hours',y='marks',hue='age',data=dataset,kind='line',col='internet')
```

## DATA DISTRIBUTION - HISTOGRAM

```python
sns.displot(dataset,x='age',binwidth=0.5,hue='internet')
```

```python
sns.displot(dataset,x='age',bins = list(range(10,21)))
```

```python
sns.displot(dataset,x='marks',col='internet',color='g')
```

## DATA DISTRIBUTIONS - BIVARIATE DISTRIBUTIONS

```python
sns.displot(dataset,x='age',y='marks',hue='internet')
```

## DATA DISTRIBUTIONS - KERNEL DENSITY DISTRIBUTION

```python
sns.displot(dataset,x='age',y='marks',kind="kde",hue='internet')
```

## PAIRPLOT

```python
sns.pairplot(dataset)
```

## CATEGORICAL DATA - SCATTERPLOT

```python
sns.catplot(x="age",y="marks",data=dataset,hue="internet")
```

## CATEGORICAL DATA - BOX PLOT

```python
sns.catplot(x="age",y="marks",data=dataset,kind="box",hue="internet")
```

## REGRESION OUTPUT

```python
sns.regplot(x="marks",y="hours",data=dataset)
```

```python
sns.lmplot(x="marks",y="hours",data=dataset,hue="internet",markers=["o","x"],palette="Set1"); # because of semi-colon object addr doesn't print
```
